```python
from scipy import misc
```

```python
asccent_image=misc.ascent()
```

```python
import matplotlib.pyplot as plt

```

```python
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(asccent_image)
plt.show()
```

```python
import numpy as np
image_transformed=np.copy(asccent_image)
```

```python
size_x = image_transformed.shape[0]
size_y = image_transformed.shape[1]
```

```python
size_x
```

```python
size_y
```

```python
# Experiment with different values and see the effect
filter = [ [0, 1, 0], [1, -4, 1], [0, 1, 0]]

# A couple more filters to try for fun!
# filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]
# filter = [ [-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]

# If all the digits in the filter don't add up to 0 or 1, you 
# should probably do a weight to get it to do so
# so, for example, if your weights are 1,1,1 1,2,1 1,1,1
# They add up to 10, so you would set a weight of .1 if you want to normalize them
weight  = 1
```

```python
# Iterate over the image
for x in range(1,size_x-1):
  for y in range(1,size_y-1):
      convolution = 0.0
      convolution = convolution + (asccent_image[x-1, y-1] * filter[0][0])
      convolution = convolution + (asccent_image[x-1, y] * filter[0][1])  
      convolution = convolution + (asccent_image[x-1, y+1] * filter[0][2])     
      convolution = convolution + (asccent_image[x, y-1] * filter[1][0])    
      convolution = convolution + (asccent_image[x, y] * filter[1][1])    
      convolution = convolution + (asccent_image[x, y+1] * filter[1][2])    
      convolution = convolution + (asccent_image[x+1, y-1] * filter[2][0])    
      convolution = convolution + (asccent_image[x+1, y] * filter[2][1])    
      convolution = convolution + (asccent_image[x+1, y+1] * filter[2][2])    
      
      # Multiply by weight
      convolution = convolution * weight   
      
      # Check the boundaries of the pixel values if less than 0 convert it to zero simmilarly for 225
      if(convolution<0):
        convolution=0
      if(convolution>255):
        convolution=255

      # Load into the transformed image
      image_transformed[x, y] = convolution
```

```python
# Plot the image. Note the size of the axes -- they are 512 by 512
plt.gray()
plt.grid(False)
plt.imshow(image_transformed)
plt.show()  
```

```python
# Assign dimensions half the size of the original image
new_x = int(size_x/2)
new_y = int(size_y/2)

# Create blank image with reduced dimensions
newImage = np.zeros((new_x, new_y))

# Iterate over the image
for x in range(0, size_x, 2):
  for y in range(0, size_y, 2):
    
    # Store all the pixel values in the (2,2) pool
    pixels = []
    pixels.append(image_transformed[x, y])
    pixels.append(image_transformed[x+1, y])
    pixels.append(image_transformed[x, y+1])
    pixels.append(image_transformed[x+1, y+1])

    # Get only the largest value and assign to the reduced image
    newImage[int(x/2),int(y/2)] = max(pixels)

# Plot the image. Note the size of the axes -- it is now 256 pixels instead of 512
plt.gray()
plt.grid(False)
plt.imshow(newImage)
plt.show()     
```
